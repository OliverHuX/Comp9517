Task3 0 degrees

```python
img = cv2.imread('syd_house.jpg')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

rows, cols = img.shape[:2]
M = cv2.getRotationMatrix2D((cols/2, rows/2), 0, 1)
rotate_img = cv2.warpAffine(img, M, (cols, rows))
gray_rotated = cv2.cvtColor(rotate_img, cv2.COLOR_BGR2GRAY)

#If u r using the 3.4.3 version
#Please change the function of creating the SIFT into the following
#sift = cv.xfeatures2d.SIFT_create()
sift = cv2.SIFT_create()
kp1, des1 = sift.detectAndCompute(gray_img, None)
kp2, des2 = sift.detectAndCompute(gray_rotated, None)


bf = cv2.BFMatcher()

matches = bf.knnMatch(des1,des2, k=2)
good = []
for m,n in matches:
    if m.distance < 0.7*n.distance:
        good.append([m])
result = cv2.drawMatchesKnn(gray_img, kp1, gray_rotated, kp2, good, None, flags=2)

plt.imshow(result)
cv2.imwrite('task3_0_result.jpg', result)
```
