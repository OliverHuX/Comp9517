```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from skimage.feature import hog
import statistics
import cv2
import os
import glob
import json
import numpy as np
import random
import time
import pickle
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
```

```python
# this function crops the image from bbox section in annotation.json
# save the cropped image in fold named cars in current path
# make sure the dataset is located at the parent path like '../benchmark_velocity_train'
def cars():
    count = 1
    dataset_path = '../benchmark_velocity_train/clips/'
    folder_path = os.listdir(dataset_path)
    for folder in folder_path:
        if folder == '.DS_Store':
            continue
        json_path = os.path.join(dataset_path, folder, 'annotation.json')
        print(json_path)
        img_folder = os.path.join(dataset_path, folder, 'imgs')
        imgs = os.listdir(img_folder)
        jf = open(json_path, encoding='utf-8')
        json_f = json.load(jf)
        for info in json_f:
            bbox = info['bbox']
            left = int(bbox['left'])
            top = int(bbox['top'])
            bottom = int(bbox['bottom'])
            right = int(bbox['right'])
            img = imgs[39]
            name = str(count)+'.jpg'
            img = os.path.join(img_folder, img)
            i = cv2.imread(img, cv2.COLOR_BGR2RGB)
            crop = i[top:bottom, left:right]
            c = cv2.resize(crop, (64,64))
            plt.imshow(c)
            if not os.path.exists('./cars'):
                os.mkdir('./cars')
            cv2.imwrite('./cars/'+name, c)
            count += 1
    
# cars()
```

```python
# generate block which is no overlap with bbox in annotation
def generate_cell(hight, width, size, car_left, car_right, car_top, car_bottom):
    flag = True
    while flag:
        count = 0
        left = random.randrange(0, width-size+1)
        right = left+size
        top = random.randrange(0, hight-size+1)
        bottom = top+size
        
        for i in range(len(car_left)):
            
            if top > car_top[i] and top < car_bottom[i] and right > car_left[i] and right < car_right[i]:
                break
            elif bottom > car_top[i] and bottom < car_bottom[i] and right > car_left[i] and right < car_right[i]:
                break
            elif bottom < car_bottom[i] and top > car_top[i] and right < car_right[i] and left > car_left[i]:
                break
            else:
                count += 1
        
        if count == len(car_left):
            flag = False
    return left, top
            
# crop image by block generate by function generate_cell()
# save the cropped image in fold named noncars in current path
# make sure the dataset is located at the parent path like '../benchmark_velocity_train'
def notcars():
    count = 1
    dataset_path = '../benchmark_velocity_train/clips/'
    folder_path = os.listdir(dataset_path)
    for folder in folder_path:
        if folder == '.DS_Store':
            continue
        json_path = os.path.join(dataset_path, folder, 'annotation.json')
        print(json_path)
        img_folder = os.path.join(dataset_path, folder, 'imgs')
        imgs = os.listdir(img_folder)
        jf = open(json_path, encoding='utf-8')
        json_f = json.load(jf)
        left = []
        right = []
        top = []
        bottom = []
        for info in json_f:
            bbox = info['bbox']
            left.append(int(bbox['left']))
            top.append(int(bbox['top']))
            bottom.append(int(bbox['bottom']))
            right.append(int(bbox['right']))
            
        img = imgs[39]
        name = str(count)+'.jpg'
        img = os.path.join(img_folder, img)
        i = cv2.imread(img, cv2.COLOR_BGR2RGB)
        h = i.shape[0]
        w = i.shape[1]
        l, t = generate_cell(h, w, 128, left, right, top, bottom)
        crop = i[t:t+128, l:l+128]
        c = cv2.resize(crop, (64,64))
        if not os.path.exists('./noncars'):
            os.mkdir('./noncars')
        cv2.imwrite('./noncars/'+name, c)
        count += 1
        
# notcars()
```

```python
print('processing car features...')
t1 = time.time()
car_imgs = glob.glob('./cars/*.jpg')
car_features = extract_features(car_imgs)
print('car features done')
t2 = time.time()
print(f'Takes {round(t2-t1)}s to process car features')
print('processing none-car features...')
noncar_imgs = glob.glob('./noncars/*.jpg')
noncar_features = extract_features(noncar_imgs)
t3 = time.time()
print(f'Takes {round(t3-t2)}s to process none-car features')
features = np.vstack((car_features, noncar_features))
label = np.hstack((np.ones(len(car_features)), np.zeros(len(noncar_features))))
x_train,x_test,y_train,y_test = train_test_split(features, label, train_size = 0.8, random_state=42)

print(f'All done, {round(t3-t1)}s in total')
```

```python
t1 = time.time()
# https://scikit-image.org/docs/dev/api/skimage.feature.html?highlight=hog#skimage.feature.hog
print('Start training...')
svc = LinearSVC()
svc.fit(x_train, y_train)
t2 = time.time()
print(f'Takes {round(t2-t1)}s to train SVC')

print('Save SVC modal')
f = open('modal', 'wb')
pickle.dump(svc,f)
f.close
print('done')
```

```python
train_accuracy = svc.score(x_train, y_train)
test_accuracy = svc.score(x_test,y_test)
print('Train Accuracy: {}'.format(round(train_accuracy,4)))
print('Test Accuracy: {}'.format(round(test_accuracy, 4)))
t = time.time()
n_predict = 100
print('My SVC predicts: ', svc.predict(x_test[0:n_predict]))
print('For these',n_predict,'labels: ',y_test[0:n_predict])
t2 = time.time()
print(round(t2-t,5),'Seconds to predict', n_predict,'labels with SVC')  
```

```python
def get_neighbors(col, row, labels):
    neighbors = []
    for i in range(0,3):
        for j in range(0,3): 
            if 0 <= col-1+i < labels.shape[0] and 0 <= row-1+j < labels.shape[1]:
                if labels[col-1+i][row-1+j] != 0:
                    neighbors.append(labels[col-1+i][row-1+j])
    return neighbors

def find_label(linked, label):
    lab = min(linked[label])
    return lab

def connected_component_labeling(image):
    background_colour = 0
    col = image.shape[0]
    row = image.shape[1]
    linked = {}
    index = 0
    labels = np.zeros((col, row), dtype=int)
    
    #First pass
    for r in range(0, row):
        for c in range(0, col):
            if image[c][r] != background_colour:
                neighbors = get_neighbors(c, r, labels)
                if not neighbors:
                    index += 1
                    linked[index] = index
                    labels[c][r] = index
                else:
                    labels[c][r] = min(neighbors)
                    for label in neighbors:
                        s = set()
                        if isinstance(linked[label], int):
                            s.add(linked[label])
                        else:
                            s = set(linked[label])
                        linked[label] = s.union(neighbors)
                        
    #Second pass
    for r in range(0, row):
        for c in range(0, col):
            if image[c][r] != background_colour:
                labels[c][r] = find_label(linked, labels[c][r])
    
    
    return labels

def search_cars(img, svc, overlap=0.9):
# convert img into gray scale
    draw = np.copy(img)
    gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    print(gray.shape)
    x_range = (0, img.shape[1])
    y_range = (0, img.shape[0])
    x = x_range[1] - x_range[0]
    y = y_range[1] - y_range[0]
    
    fig, ax = plt.subplots()
    ax.imshow(img)
    top = bottom = left = right = 0
# frequence is the heat map
    frequence = np.zeros(gray.shape)
    boxes = []
    
    while True:
# This is the scalable windows
        left = right = 0
        if top < 300:
            window_size = (300, 300)
        elif top < 400:
            window_size = (80, 80)
        elif top < 500:
            window_size = (200, 200)
        elif left < 400:
            window_size = (300, 300)
        else:
            window_size = (300, 300)
# skip changing top value in first loop
        if bottom != 0:
            top = top + int(window_size[0]*(1-overlap))
        bottom = top + window_size[0]

        if bottom > y:
            break
        while True:
# skip changing right value in first loop
            if right != 0:
                left = left + int(window_size[1]*(1-overlap))
            right = left + window_size[1]
    
            if right > x:
                break
# crop the image
# apply the hog to extract the features
            crop = gray[top:bottom, left:right]
            crop_img = cv2.resize(crop, (64,64))
            hog_features = hog(crop_img, orientations=9, pixels_per_cell=(8, 8), cells_per_block=(8, 8),
                              transform_sqrt=True, visualize=False, feature_vector=True)
            hog_features = hog_features.reshape(1,-1)
#             using the svc model to detect
            result = svc.predict(hog_features)
            if result:
                boxes.append(((left, top),(right, bottom)))
# produce the heat map
    for box in boxes:
        frequence[box[0][1]:box[1][1], box[0][0]:box[1][0]] += 1
#     plt.imshow(frequence, cmap = 'gray')
#     plt.show()   
    for r in range(frequence.shape[0]):
        for c in range(frequence.shape[1]):
            if frequence[r][c] <= 7:
                frequence[r][c] = 0
#     plt.imshow(frequence, cmap = 'gray')
#     plt.show()
# apply erosion to reduce noise
    kernel = np.ones((5,5),np.uint8)
#     https://docs.opencv.org/3.4/db/df6/tutorial_erosion_dilatation.html
    erosion = cv2.erode(frequence,kernel,iterations = 1)
# apply ccl to label each area
    labels_im = connected_component_labeling(erosion)

    cars = []
    for pixel in labels_im.ravel():
        if pixel != 0 and pixel not in cars:
            cars.append(pixel)
    
#     plt.imshow(labels_im)
#     plt.show()
# get the location of each car
    for car in cars:    
        arr = np.where(labels_im == car)
        top = min(arr[0])
        bottom = max(arr[0])
        left = min(arr[1])
        right = max(arr[1])
#         https://stackoverflow.com/questions/37435369/matplotlib-how-to-draw-a-rectangle-on-image
#         rect = patches.Rectangle((left, top*0.9), right-left, bottom-top, linewidth=3, 
#                                  edgecolor='green', facecolor='none')
#         ax.add_patch(rect)
#     plt.show()
        draw = cv2.rectangle(draw, (left, int(top*0.9), right-left, bottom-top), (0, 255, 0), 5)
    return draw

    
test_img = cv2.imread('../benchmark_velocity_train/clips/1/imgs/001.jpg')
plt.imshow(test_img)
plt.show()
draw = search_cars(test_img, svc, 0.9)
plt.imshow(draw)
plt.show()
```
